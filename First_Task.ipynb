```python
import glob
import os
import cv2
import colorsys
import pandas as pd
```

```python
img_dir = "/home/yash/Desktop/challenge/challenge/" 
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
col= ['Name','Height','Width','Average lightness','Average saturation','[0/12 - 1/12]','[1/12 - 2/12]','[2/12 - 3/12]','[3/12 - 4/12]','[4/12 - 5/12]','[5/12 - 6/12]','[6/12 - 7/12]','[7/12 - 8/12]','[8/12 - 9/12]','[9/12 - 10/12]','[10/12 - 11/12]','[11/12 - 12/12]']

    
submission = pd.DataFrame(columns = col)     #Defining the dataframe
d = []

for f1 in files:                             #Looping through the files
    img = cv2.imread(f1)
    
    sum_lightness = 0
    sum_saturation = 0
    hue_pix = [0]*12
    
    #Looping through each pixel
    for row in range(0,img.shape[0]):
        for column in range(0,img.shape[1]):
            rgb = img[row,column]
            
            #converting RGB to HLS
            hls = colorsys.rgb_to_hls(rgb[0]/255.0,rgb[1]/255.0,rgb[2]/255.0)
            
            #summing the lightness value
            sum_lightness = sum_lightness + hls[1]
            
            #summing the saturation value
            sum_saturation = sum_saturation + hls[2]  
            
            
            #calculating total no. of pixels lying in differnt range
            if hls[0]>= 0/12 and hls[0]< 1/12:
                hue_pix[0] = hue_pix[0] + 1
            elif hls[0]>= 1/12 and hls[0]< 2/12:
                hue_pix[1] = hue_pix[1] + 1
            elif hls[0]>= 2/12 and hls[0]< 3/12:
                hue_pix[2] = hue_pix[2] + 1
            elif hls[0]>= 3/12 and hls[0]< 4/12:
                hue_pix[3] = hue_pix[3] + 1
            elif hls[0]>= 4/12 and hls[0]< 5/12:
                hue_pix[4] = hue_pix[4] + 1
            elif hls[0]>= 5/12 and hls[0]< 6/12:
                hue_pix[5] = hue_pix[5] + 1
            elif hls[0]>= 6/12 and hls[0]< 7/12:
                hue_pix[6] = hue_pix[6] + 1
            elif hls[0]>= 7/12 and hls[0]< 7/12:
                hue_pix[7] = hue_pix[7] + 1
            elif hls[0]>= 8/12 and hls[0]< 7/12:
                hue_pix[8] = hue_pix[8] + 1
            elif hls[0]>= 9/12 and hls[0]< 7/12:
                hue_pix[9] = hue_pix[9] + 1
            elif hls[0]>= 10/12 and hls[0]< 11/12:
                hue_pix[10] = hue_pix[10] + 1
            elif hls[0]>= 11/12 and hls[0]< 12/12:
                hue_pix[11] = hue_pix[11] + 1
                
    #list          
    d = [[f1.split('/')[6],img.shape[0],img.shape[1],sum_lightness/(img.shape[0]*img.shape[1]),sum_saturation/(img.shape[0]*img.shape[1]),hue_pix[0],hue_pix[1],hue_pix[2],hue_pix[3],hue_pix[4],hue_pix[5],hue_pix[6],hue_pix[7],hue_pix[8],hue_pix[9],hue_pix[10],hue_pix[11]]]
   
    #dataframe for the info of the image file
    df = pd.DataFrame(data = d,columns = col)
    
    #Appending the file information in submission dataframe
    submission = submission.append(df,ignore_index =True)  

                
    
            
            
            
    
    
    
    
    
    
```
